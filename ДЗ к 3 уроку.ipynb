```python
#среднее арифметическое
ar_mean = sum(z)/20
print(ar_mean)
```

```python
#cреднее квадратичное отклонение
x = (z - ar_mean)**2 
print(x) #массив отклонений от среднего арифметического
sigma = (x.sum()/20)**0.5
sigma_1 = (x.sum()/20-1)**0.5
print(sigma, sigma_1)
```

```python
#дисперсия
variance = x.sum()/20
variance_1 = x.sum()/(20-1)
print(variance, variance_1)
```
